```python
def checkPass(p) :
  length = digit = upper = lower = False
  if len(p) < 8 :
    return False
  for i in range(len(p)) :
    if '0' <= p[i] <= '9' :
      digit = True
      continue
    elif 'a' <= p[i] <= 'z' :
      lower = True
      continue
    elif 'A' <= p[i] <= 'Z' :
      upper = True
  return digit and upper and lower

while True :
  passWord = input("Enter your password : ")
  if checkPass(passWord) :
    print("사용할 수 있습니다.")
    break
  else :
    print("사용할 수 없습니다. 다시 입력하세요.")
```
